Top Movies from 1927 to 2021

```python
i = 1927
movie_years_list = []
while i < 2022:
    url = 'https://en.wikipedia.org/wiki/{i}_in_film'.format(i=i)
    df = pd.read_html(url)
    if i == 1996:
        movies = df[4]
    else:
        movies = df[3]
    movies['Year'] = i
    movies.columns = ['Rank', 'Title', 'Studio', 'Gross', 'Year']
    movie_years_list.append(movies)
    i = i + 1

movie_years_list
all_movies = pd.concat(movie_years_list)
all_movies
```

```python
i = 1970
movie_decades_list = []
while i < 2021:
    url = 'https://en.wikipedia.org/wiki/{i}s_in_film'.format(i=i)
    df = pd.read_html(url)
    if i == 1970:
        movies = df[4]
    elif i == 1980:
        movies = df[6]
    else:
        movies = df[3]
    
    movies.drop(movies.columns[5],axis=1, inplace=True)
    # movies.drop('Ref', axis=1)
    movies.columns = ['Rank', 'Title', 'Studio', 'Gross', 'Year']
    movie_decades_list.append(movies)
    i = i + 10

movie_decades_list
all_decades = pd.concat(movie_decades_list)
all_decades
```

```python
all_movies = pd.concat([all_movies, all_decades])
all_movies.drop_duplicates(subset =['Title', 'Year'], keep = 'first', inplace = True)
all_movies
```

```python
def fix_gross(str):
    # Remove text between []
    no_desc = re.sub("([\(\[]).*?([\)\]])", "\g<1>\g<2>", str)

    # Check characters to see if they are in punctuation
    nopunc = [char for char in no_desc if char not in string.punctuation]

    # Join the characters again to form the string.
    nopunc = ''.join(nopunc)

    num = int(nopunc)
    return num

all_movies['Gross'] = all_movies['Gross'].apply(fix_gross)
all_movies.reset_index(drop=True, inplace=True)
all_movies
```

## Parse Words from SRT files

based on [pablo-var/learn-english-words-from-srt](https://github.com/pablo-var/learn-english-words-from-srt) repository

```python
def is_time_stamp(l):
  if l[:2].isnumeric() and l[2] == ':':
    return True
  return False

def has_letters(line):
  if re.search('[a-zA-Z]', line):
    return True
  return False

def remove_non_ascii(text):
    return ''.join(i for i in text if ord(i)<128)

def has_no_text(line):
  l = line.strip()
  if not len(l):
    return True
  if l.isnumeric():
    return True
  if is_time_stamp(l):
    return True
  if l[0] == '(' and l[-1] == ')':
    return True
  if not has_letters(line):
    return True
  return False

def is_lowercase_letter_or_comma(letter):
  if letter.isalpha() and letter.lower() == letter:
    return True
  if letter == ',':
    return True
  return False

def clean_up(lines):
  """
  Get rid of all non-text lines and
  try to combine text broken into multiple lines
  """
  new_lines = []
  for line in lines[1:]:
    line = remove_non_ascii(line)
    if has_no_text(line):
      continue
    elif len(new_lines) and is_lowercase_letter_or_comma(line[0]):
      #combine with previous line
      new_lines[-1] = new_lines[-1].strip() + ' ' + line
    else:
      #append line
      new_lines.append(line)
  return new_lines

def main(args):
  """
    args[1]: file name
    args[2]: csv name
    args[3]: encoding. Default: utf-8.
      - If you get a lot of [?]s replacing characters,
      - you probably need to change file_encoding to 'cp1252'
  """
  file_name = args[1]
  file_encoding = 'utf-8' if len(args) < 4 else args[3]

  with open(file_name, encoding=file_encoding, errors='replace') as f:
    lines = f.readlines()
    new_lines = clean_up(lines)
  
  new_file_name = file_name[:-4] + '.txt'

  with open(new_file_name, 'w') as f:
    for line in new_lines:
      f.write(line)
  
  with open(new_file_name, 'r') as myfile:
    data = re.sub('\W+',' ', myfile.read()).replace('\n', '').lower()
  
  stop_words = set(stopwords.words("english"))  # load stopwords

  words = [w for w in data.split(" ") if w not in stop_words]

  new_data = Counter(words)

  for key, count in dropwhile(lambda key_count: key_count[1] >= 2, new_data.most_common()):
    del new_data[key]
  
  df = pd.DataFrame.from_dict(new_data.most_common())
  df.columns = ['word', 'occurrences']
  csv_name = args[2]
  df.to_csv(csv_name, index = False)

prob_movies = []

for file_name in os.listdir("srts"):
    
    with open(os.path.join("srts", file_name), 'r', encoding="ISO-8859-1") as f:
       lines = f.readlines()
       new_lines = clean_up(lines)
    
    new_file_name = file_name[:-4] + '.txt'

    with open(new_file_name, 'w') as f:
      for line in new_lines:
        f.write(line)
    
    with open(new_file_name, 'r') as myfile:
      data = re.sub('\W+',' ', myfile.read()).replace('\n', '').lower()

    stop_words = set(stopwords.words("english"))  # load stopwords

    words = [w for w in data.split(" ") if w not in stop_words]

    new_data = Counter(words)

    for key, count in dropwhile(lambda key_count: key_count[1] >= 2, new_data.most_common()):
      del new_data[key]

    movie_strs = file_name[:-4].split('-')
    l = len(movie_strs)
    movie_name = ' '.join(movie_strs[0:l-1])
    movie_year = int(movie_strs[l-1])
    movie_first_word = movie_strs[0]
    movie_last_word = movie_strs[l-2]
    # print(movie_name, movie_year, dict(new_data.most_common()))

    found_movies = all_movies[all_movies['Title'].apply(lambda x: x.lower().__contains__(movie_last_word.lower())) & (all_movies['Title'].apply(lambda x: x.lower().__contains__(movie_first_word.lower())) ) & (all_movies['Year'] == movie_year)]

    i = found_movies.index
    
    
    if len(i) == 1:
      # all_movies.loc[all_movies.index[i[0]], 'Words'] = dict(new_data.most_common())
      m = all_movies.index[i[0]]
      all_movies.loc[m, 'Words'] = [deepcopy(dict(new_data.most_common()))]
      all_movies.loc[all_movies.index[i[0]], 'Word Counts'] = len(new_data.most_common())
    else:
      prob_movies.append({'movie': movie_name, 'year': movie_year})
      

```
